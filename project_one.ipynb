### Код для проекта

```python
#библиотеки, используемые для проекта

import os
import pandas as pd
from colorsys import rgb_to_hsv
from collections import defaultdict
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns

```

Работа с категорией "техника"

```python
# функция определения доминантного цвета на
def get_dominant_color(image_path):
 original = Image.open(image_path)
 reduced = original.convert("P", palette=Image.WEB)
 palette = reduced.getpalette()
 palette = [palette[3*n:3*n+3] for n in range(256)]
 color_count = [(n, palette[m]) for n,m in reduced.getcolors()]
 
  # Исключение белого и сероватого цветов
 color_count = [color for color in color_count if color[1] != [255, 255, 255] and color[1] != [128, 128, 128]]
 
 if color_count:
   return max(color_count, key=lambda x: x[0])[1]
 else:
   return None

def get_color_group(color):
 # Здесь вы можете определить, к какой группе цветов относится каждый цвет
 # Например, можно использовать определенные диапазоны значений RGB
 
 if color[0] < 200 and color[1] < 128:
     return 'brown'
 elif color[0] < 128 and color[1] < 128:
     return 'red'
 elif color[0] < 128 and color[2] < 128:
     return 'blue'
 elif color[1] < 128 and color[2] < 128:
     return 'green'
 elif color[0] < 128 and color[1] < 128 and color[2] < 128:
     return 'yellow'
 elif color[0] > 128 and color[1] < 128 and color[2] < 128:
    return 'orange'
 elif color[0] < 128 and color[1] > 128 and color[2] < 128:
    return 'purple'
 else:
     return 'other'
 
 
# Путь к директории с изображениями
image_dir = 'screenshots\machinery'

# Получение списка всех файлов в директории
image_files = os.listdir(image_dir)

# Словарь для подсчета доминантных цветов
color_counts = defaultdict(int)
data = [['image_path', 'dominant_color']]
# Применение функции get_dominant_color к каждому файлу
array_image_path = []
array_dominant_color = []
array_color_name = []
for image_file in image_files:
    image_path = os.path.join(image_dir, image_file)
    dominant_color = get_dominant_color(image_path)
    if dominant_color is not None:
        color_group = get_color_group(dominant_color)
        color_counts[color_group] += 1
    array_image_path.append(image_path)
    array_color_name.append(color_group)
    array_dominant_color.append(dominant_color)
    current_data = [image_path, dominant_color]
    data.append(current_data)

df = pd.DataFrame({"image_path": array_image_path, "dominant_color": array_dominant_color , "color_name": array_color_name })

```

```python
print(df)
```

```python
df['color_name']
```

```python
df['color_name'].value_counts()
```

```python
df.color_name.value_counts().plot.pie(colors=[ "grey", "brown", "green", "blue"]).legend(labels=["other","brown","green","blue"], loc='center right', bbox_to_anchor=(1.8, 0.5)).set_title("Цвета категории Техника")
```
